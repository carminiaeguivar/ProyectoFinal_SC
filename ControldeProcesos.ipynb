```python
import pandas as pd
import numpy
from pyspc import *
df = pd.read_csv('ozone.csv')
df.head()
# dataset source: https://vincentarelbundock.github.io/Rdatasets/datasets.html
# Monthly provisional mean total ozone (in Dobson units) at Halley Bay (approximately corrected to Bass-Paur).
# From 1956 to 2000
# August to April :6 continuous months
```

```python
# Depuramos algunas columnas
```

```python
df=df.drop(['nro','Year','Annual'], axis=1)
df.head()
```

```python
a = spc(df) + ewma()
print(a)
```

```python
a + cusum() + xbar_sbar() + sbar()
print(a)
```

```python
a = spc(df) + xbar_rbar() + rbar() + rules()
print(a)
```
